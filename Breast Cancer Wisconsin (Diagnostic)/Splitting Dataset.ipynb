## **Splitting the Classification Dataset**

```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('data.csv')
df = df.drop(df.columns[-1], axis = 1)
df
```

```python
# shuffle the DataFrame rows & return all rows
# df1 = df.sample(frac = 1).reset_index(drop = True)
# df1
```

```python
train, test = train_test_split(df, test_size = 0.2, random_state = 420)
```

```python
train = train.reset_index(drop = True)
train
```

```python
test = test.reset_index(drop = True)
test
```

```python
train.to_csv('Classification - Training.csv')
test.to_csv('Classification - Testing.csv')
```
